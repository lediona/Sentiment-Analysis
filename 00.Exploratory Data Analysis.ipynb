#### Note:  In conclusion of inspecting missing values: we have: 
    -13621 Null values  in the tweet_coord column: empty coordinates
    -164 tweet_cord instances that have the coordinates [0.0, 0.0]
    -14640 total number of rows from airline_tweets_df
